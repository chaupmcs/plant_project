## Up

```python
heatmap_up_data = df_up[["atID"]].copy()
for c in stop1_cols[1:]:
    tmp = np.sqrt(np.abs(df_up[c] - df_up[stop1_cols[0]])) 
    heatmap_up_data[c] = tmp * np.sign(df_up[c] - df_up[stop1_cols[0]])

heatmap_up_data['tmp'] = heatmap_up_data.iloc[:, 1:].sum(axis=1)
    
heatmap_up_data = heatmap_up_data.reset_index(drop=True)
heatmap_up_data.sort_values('tmp', inplace=True)
heatmap_up_data.head()
```

```python
import plotly.graph_objects as go
import datetime
import numpy as np
np.random.seed(1)

fig = go.Figure(data=go.Heatmap(
        z=heatmap_up_data.iloc[:,1:],
        zmid = 0,
        x=stop1_cols[1:],
        y=heatmap_up_data['atID'],
        colorscale=coolwarm))

fig.update_layout(
    title='Upregulated (wt - treated)',
    xaxis_nticks=36)

fig.show()
```
